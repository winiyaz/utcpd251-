The formula for the average (arithmetic mean) of a set of numbers $x_1, x_2, \ldots, x_n$ is given by:

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

where:
- $\bar{x}$ is the average,
- $n$ is the number of observations,
- $x_i$ represents each individual observation.

```python
# Average will be like this from df 
te_li = df['temp']
rprint(te_li)
inspect(te_li)
```

```python
av1 = sum(te_li) / len(te_li)
rprint(f'''
Average = {av1}
Sum = {sum(te_li)}
Len = {len(te_li)}
''')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'temp': [12, 14, 15, 14, 21],
    'condition': ['Sunny', 'Rain', 'Rain', 'Cloudy', 'Sunny']
}
df = pd.DataFrame(data)

# Select only the numerical columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
corr_matrix = numerical_df.corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
