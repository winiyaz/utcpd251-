## Given solution

```python
# Given solution 

sqdf_c1 = sqdf[sqdf['Primary Fur Color'] == 'Gray']
```

```python
# geting the counts 
sqdf_grey_count = len(sqdf_c1)
rprint(sqdf_grey_count)
```

```python
# Doing all 3 in the shown method 
sqdf_G = len(sqdf[sqdf['Primary Fur Color'] == 'Gray'])
sqdf_C = len(sqdf[sqdf['Primary Fur Color'] == 'Cinnamon'])
sqdf_B= len(sqdf[sqdf['Primary Fur Color'] == 'Black'])

rprint(f'''
Squirrel Counts 

Grey = {sqdf_G}
Cina = {sqdf_C}
Black = {sqdf_B}
''')
```

```python
# Constructing the data frame using dictionary method 
sqdf_c_di = {
    "FurColor" : ["Gray", "Cina", "Blak"],
    "Count" : [sqdf_G, sqdf_C, sqdf_B]
}

pprint(sqdf_c_di, expand_all=True)
```

```python
# Creating the DF now
sqdf2C = pd.DataFrame(sqdf_c_di)
```
